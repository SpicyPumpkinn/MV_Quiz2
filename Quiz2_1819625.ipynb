```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image = cv2.imread('lego1.png')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Plot the grayscale image
plt.figure()
plt.imshow(gray, cmap='gray')
plt.title('Grayscale image')
```

```python
# Binarize the image (converting it into a black and white (two-color) image)
_, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
```

```python
# Plot the binary image
plt.imshow(binary, cmap='gray')
plt.title('Binary image')
plt.show()
```

```python
# Perform morphological operations to remove noise
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, iterations=2)
```

```python
# Plot the image after morphological opening
plt.imshow(opening, cmap='gray')
plt.title('Image after morphological opening')
plt.show()
```

```python
# Perform morphological closing to fill in small holes
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel, iterations=3)
```

```python
# Plot the image after morphological closing
plt.imshow(closing, cmap='gray')
plt.title('Image after morphological closing')
plt.show()
```

```python
# Erode away small circles inside the objects
erosion = cv2.erode(closing, kernel, iterations=3)
```

```python
# Plot the image after erosion
plt.imshow(erosion, cmap='gray')
plt.title('Image after erosion')
plt.show()
```

```python
# Perform edge detection
edges = cv2.Canny(erosion, 50, 100)
```

```python
# Plot the edges
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.show()
```

```python
# Find the contours of the edges
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Plot the contours on the original image
image_contours = cv2.drawContours(image, contours, -1, (0,0,255), 2)
plt.imshow(image_contours)
plt.title('Contours on original image')
plt.show()
```

```python
# Print the number of objects in the image
print(f'Number of objects in the image: {len(contours)}')
```

```python
""" 
The image is read and converted to grayscale, to simplify the image and reduce the number of color channels.

The grayscale image is binarized using a threshold value, to convert the image into a black and white (two-color) image.

Morphological opening is applied to the binary image, to remove small noise and details from the image.

Morphological closing is applied to the image after morphological opening, to fill in small holes in the objects.

Erosion is applied to the image after morphological closing, to erode away small circles inside the objects.

The edges of the objects are detected using the Canny edge detector.

The contours of the edges are found using the cv2.findContours() function.

The number of contours is counted, and this is the number of objects in the image.
"""
```
